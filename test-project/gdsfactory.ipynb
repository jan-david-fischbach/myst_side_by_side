```python
#| label: geometry-polygon

import gdsfactory as gf
from gdsfactory.gpdk import PDK

PDK.activate()

# Create a blank component 
# (essentially an empty GDS cell 
# with some special features).
c = gf.Component()
p1 = c.add_polygon(
    [(-8, -6), (6, 8), (7, 17), (9, 5)], 
    layer=(1, 0)
)

# Write it to a GDS file.
c.write_gds("demo.gds")  
c.show()  # Show it in klayout.
c.plot();  # Plot it in jupyter notebook.
```
